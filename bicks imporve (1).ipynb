```python
import requests
from bs4 import BeautifulSoup
import pandas as pdbbb
```

```python
page = requests.get("https://www.bikewale.com/new-bike-search/?bodyStyleIds=5")
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
page.status_code
```

```python
rating_tags = soup.select(".o-eemiLE.o-bdcqVx.o-cKuOoN.o-lIIwF.o-eZTujG")
rating = []
for tag in rating_tags:
    text = tag.get_text()
    cleaned_text = text.replace('/5', '').strip()  # Remove the '/5' part and extra whitespace
    rating.append(cleaned_text)
    print("Ratings:", rating)
```

```python
len(rating)
```

```python
price_tags = soup.select(".o-eZTujG.o-byFsZJ.o-bkmzIL.o-bVSleT")
price = []
for tag in price_tags:
    text = tag.get_text()
    cleaned_text = text.replace('₹', '').strip()  # Remove the Rupee symbol and extra whitespace
    price.append(cleaned_text)
    print("Prices:", price)
```

```python
len(price)
```

```python
Bike_tags=soup.select(".o-eZTujG.o-bkmzIL.o-eqqVmt.o-fzpibr")
Bike=[]
for i in Bike_tags:
    Bike.append(i.get_text())
print(Bike)
```

```python
len(Bike)
```

```python
avg_tags=soup.select(".o-bqHweY.o-bVSleT.o-bwCunT.o-cgFpsP.o-bNxxEB.o-cpNAVm.o-eemiLE")
avg=[]
for i in avg_tags:
    avg.append(i.get_text())
print(avg)
```

```python
len(avg)
```

```python
import pandas as pd
Bikes = pd.DataFrame({
    "Bike Name":Bike,
    "Bike rating":rating,
    "Bike price": price,
    "Bike avg":avg
})
Bikes
```

```python
Bikes.to_csv("Bikes_Report_Scraping.csv",index=False)
```
